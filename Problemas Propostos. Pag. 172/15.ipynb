<b>Traçar um esboço do gráfico e obter uma equação da parábola que satisfaça as condições dadas.</b>

<b>15. Focos: $F(0,-\frac{1}{4})$; diretriz $d: 4y -1 = 0$</b>

<b>Arrumando a equação da diretriz</b><br><br>
$d: y = \frac{1}{4}$<br><br>
<b>Se $p$ é a distância entre o foco e a diretriz, temos que  $p = -\frac{1}{2}$</b><br><br>
$F = \frac{p}{2}$<br><br>
$-\frac{1}{4} = \frac{p}{2}$<br><br>
$-\frac{1}{2} = p$<br><br><br>
<b>Montando a equação</b><br><br>
$x^2 = 2py$<br><br>
$x^2 = 2 \cdot (-\frac{1}{2}) \cdot y$<br><br>
$x^2 = -y$<br><br><br>
<b>Gráfico da função</b><br><br>

```python
from sympy import *
from sympy.plotting import plot_implicit
x, y = symbols("x y")
plot_implicit(Eq((x-0)**2, -1*(y+0)), (x,-10,10), (y,-10,10),
title=u'Gráfico da parábola', xlabel='x', ylabel='y');
```
