<b>12. Foco: $F(2,0)$; Firetriz $d: x+2=0$</b><br><br>
<b>Arrumando a equação da diretriz</b><br><br>
$x = -2$<br><br>
<b>Se a diretriz é dada por $-\frac{p}{2}$ e possui o valor $x = -2$ e o foco possui o valor $2$ para $x$ e sabendo que a distância do foco para o vértice e do vértice para a diretriz são iguais, temos que $V(0,0)$, para que isso seja verdadeiro. </b><br><br>
$D : x = -2$<br><br>
$F(2,0)$<br><br><br>
<b>Descobrindo o valor de $p$</b><br><br>
$F = \frac{p}{2}$<br><br>
$F = 2$<br><br>
$\frac{p}{2} = 2$<br><br>
$p = 4$<br><br>
<b>Montando a equação</b><br><br><br>
$y^2 = 2px$<br><br>
$y^2 = 2 \cdot 4 \cdot x$<br><br>
$y^2 = 8x$<br><br><br>
<b>Gráfico da parábola</b><br><br>

```python
from sympy import *
from sympy.plotting import plot_implicit
x, y = symbols("x y")
plot_implicit(Eq((y-0)**2, 8*(x+0)), (x,-10,10), (y,-10,10),
title=u'Gráfico da parábola', xlabel='x', ylabel='y');
```
