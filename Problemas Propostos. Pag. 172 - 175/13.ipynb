<b>Traçar um esboço do gráfico e obter uma equação da parábola que satisfaça as condições dadas.</b>

<b>13. Vértice: $V(0,0)$; Foco: $F(0,-3)$</b><br><br>


<b>Se a distância do vértice ao o foco é igual a distância do vértice até a diretriz, temos que $||F|| = ||D||$, logo $3 = d$</b><br><br><br>
<b>Equação da diretriz</b>
$D : y = 3$<br><br>
<b>Descobrindo o valor de $p$</b><br><br>
$F = -3$<br><br>
$-3 = \frac{p}{2}$<br><br>
$-6 = p$<br><br>
<b>Montando a equação</b><br><br>
$x^2 = 2py$<br><br>
$x^2 = 2\cdot -6 \cdot x$<br><br>
$x^2 = -12y$<br><br><br>
<b>Gráfico da função</b><br>

```python
from sympy import *
from sympy.plotting import plot_implicit
x, y = symbols("x y")
plot_implicit(Eq((x-0)**2, -12*(y+0)), (x,-10,10), (y,-10,10),
title=u'Gráfico da parábola', xlabel='x', ylabel='y');
```
