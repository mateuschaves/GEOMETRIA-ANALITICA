<b>Traçar um esboço do gráfico e obter uma equação da parábola que satisfaça as condições dadas.</b>

<b>11. Vértice : $V(0,0)$; diretriz $d: y = -2$</b><br><br><br>
<b>Se a diretriz é dada por $-\frac{p}{2}$ e possui o valor $y = -2$ é fácil perceber que o foco que é dado por $\frac{p}{2}$ será $2$, logo $F(0,2)$</b><br><br>
<b>Achando o valor de $p$</b><br><br>
$2 = \frac{p}{2}$<br><br>
$4 = p$<br><br>
<b>Achando a equação da parábola</b><br><br>
$x^2 = 2py$<br><br>
$x^2 = 2\cdot 4\cdot y$<br><br>
$x^2 = 8y$<br><br><br>
<b>Gráfico da parábola</b><br><br>

```python
from sympy import *
from sympy.plotting import plot_implicit
x, y = symbols("x y")
plot_implicit(Eq((x-0)**2, 8*(y+0)), (x,-10,10), (y,-10,10),
title=u'Gráfico da parábola', xlabel='x', ylabel='y');
```
