<b>Construir o gráfico e encontrar o foco e uma equação da diretriz.</b>

<b>8. $2y^2 -9x = 0$</b><br><br>
<b>Arrumando a equação</b><br><br>
$2y^2 = 9x$<br><br>
$y^2 = \frac{9x}{2}$<br><br>
$2p = \frac{9}{2}$, logo<br><br>
$p = \frac{9}{4}$<br><br><br>
<b>Calculando o foco</b><br><br>
$F = \frac{p}{2}$<br><br>
$F = \frac{\frac{9}{4}}{2}$<br><br>
$F = \frac{9}{4} \cdot \frac{1}{2}$<br><br>
$F(\frac{9}{8}, 0)$<br><br>
<b>Calculando a diretriz</b><br><br>
$D = -\frac{p}{2}$<br><br>
$D = -\frac{9}{8}$<br><br>
$V(0,0)$<br><br>
$D : x = -\frac{9}{8}$<br><br>
$F(\frac{9}{8}, 0)$<br><br>

```python
from sympy import *
from sympy.plotting import plot_implicit
x, y = symbols("x y")
plot_implicit(Eq((y-0)**2, 4.5*(x+0)), (x,-10,10), (y,-10,10),
title=u'Gráfico da parábola', xlabel='x', ylabel='y');
```
