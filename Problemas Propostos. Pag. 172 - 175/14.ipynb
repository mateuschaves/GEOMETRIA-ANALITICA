<b>14. Vértice: $V(0,0)$; Foco: $F(-\frac{1}{2},0)$</b><br><br><br>
<b>Se a distância do vértice ao foco e do vértice a diretriz são iguais, temos que se a distância do vértice para o foco é de $-\frac{1}{2}$ a diretriz é $x = \frac{1}{2}$</b>

<b>Achando o valor de $p$</b><br><br>
$F = \frac{p}{2}$<br><br>
$-\frac{1}{2} = \frac{p}{2}$<br><br>
$-1 = p$<br><br><br><br>
<b>Montando a fórmula</b><br><br>
$y^2 = 2px$<br><br>
$y^2 = 2 \cdot -1 \cdot x$<br><br>
$y^2 = -2x$<br><br><br>
<b>Gráfico da parábola</b><br><br>

```python
from sympy import *
from sympy.plotting import plot_implicit
x, y = symbols("x y")
plot_implicit(Eq((y-0)**2, -2*(x+0)), (x,-10,10), (y,-10,10),
title=u'Gráfico da parábola', xlabel='x', ylabel='y');
```
